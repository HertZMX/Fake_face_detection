```python
def check_folder_size(source_folder):
    # Get a list of all subfolders in the source folder
    subfolders = [f.name for f in os.scandir(source_folder) if f.is_dir()]

    # Loop through all subfolders
    for subfolder in subfolders:
        # Get the path to the subfolder
        subfolder_path = os.path.join(source_folder, subfolder)

        # Get the list of image files in the subfolder
        image_files = [f for f in os.listdir(subfolder_path) if f.endswith(".jpg") or f.endswith(".jpeg") or f.endswith(".png")]

        # Print the number of images in the subfolder
        print(f"Number of images in {subfolder}: {len(image_files)}")

        # Check if the subfolder has any subfolders
        sub_subfolders = [f.name for f in os.scandir(subfolder_path) if f.is_dir()]
        if sub_subfolders:
            # If the subfolder has subfolders, recursively call this function on each subfolder
            print(f"Checking subfolders in {subfolder}...")
            subfolder_path = os.path.join(source_folder, subfolder)
            check_folder_size(subfolder_path)
```

```python
def check_duplicate_filenames(folder_path):
    """Check for duplicate filenames in a folder of images.
    
    Parameters:
    folder_path (str): The path to the folder containing the images.
    
    Returns:
    duplicates (list): A list of duplicate filenames.
    """
    filenames = os.listdir(folder_path)
    duplicates = []
    seen = set()
    for filename in filenames:
        if filename in seen:
            duplicates.append(filename)
        else:
            seen.add(filename)
    return duplicates
```

```python
def check_non_numeric_filenames(folder_path):
    """Check for image files in a folder that are not named with a number.
    
    Parameters:
    folder_path (str): The path to the folder containing the images.
    
    Returns:
    non_numeric_files (list): A list of filenames that are not named with a number.
    """
    non_numeric_files = []
    for filename in os.listdir(folder_path):
        # check if the filename is a valid image file
        if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif')):
            # check if the filename contains any non-numeric characters
            if not filename[:-4].isdigit():
                non_numeric_files.append(filename)
    return non_numeric_files
```

```python
def delete_non_numeric_filenames(folder_path):
    """Delete image files in a folder that are not named with a number.
    
    Parameters:
    folder_path (str): The path to the folder containing the images.
    
    Returns:
    non_numeric_files (list): A list of filenames that were deleted.
    """
    non_numeric_files = []
    for filename in os.listdir(folder_path):
        # check if the filename is a valid image file
        if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif')):
            # check if the filename contains any non-numeric characters
            if not filename[:-4].isdigit():
                non_numeric_files.append(filename)
                os.remove(os.path.join(folder_path, filename))
    return non_numeric_files
```
